```python
### RANGE 
for i in range(10):
    print(i)
    
```

```python
#### range(start, stop)

for i in range(1, 10):
    print(i)
```

```python
#### range(start, stop, step)

for i in range(1, 30,5):
    print(i)
```

```python
##### Using range() to create a list
i = list(range(10))
print(i)
```

```python
#### if statement
age = 20

if age >= 18:
    print("You are an adult")
```

```python
#### if ,else statement

age = 16

if age >= 18:
    print("Adult")
else:
    print("Minor")
```

```python
marks = 100

if marks <= 40:
    print("pass")

else:
    print("Fail")
```

```python
### nested if
age = 25
country = "India"

if country == "India":
    if age >= 18:
        print("You can vote")
```

```python
grade =80
qualification = "masters"

if qualification == "masters":
    if grade >= 60:
        print("eligible for position")
```

```python
#### break 
for i in range(10):
    if i == 5:
        break
    print(i)
```

```python
#### continue  
for i in range(10):
    if i == 5:
        continue
    print(i)
```

```python
i = 1

while i <= 10:
    if i == 7:
        break
    print(i)
    i += 1
```

```python
##### Branching in Loops 
for i in range(1, 6):
    if i == 3:
        print("Three found!")
    else:
        print(i)
```
