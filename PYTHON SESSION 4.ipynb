```python
#### discard() vs remove()
###----- remove()
## Removes an element
### Gives an ERROR if the element is not present
### ------discard() 
##Removes an element
### DOES NOT give an error if the item is not present
```

```python
##if the item is available
products = {1, 2, 2, 3, 3, 3}
products.remove(3)
print(products)
```

```python
### if the item is not available 
products = {1, 2, 2, 3, 3, 3}
products.remove(5)
print(products)
```

```python
#### discard if the item is available
products = {1, 2, 2, 3, 3, 3,4,5,6,7,8}
products.discard(5)
print(products)
```

```python
### discard if the item is not available
products = {1, 2, 2, 3, 3, 3,4,5,6,7,8}
products.discard(10)
print(products)
```

```python
##### shallow copy
###Changes in shallow copy  affects original.
original = [1, 2, [3, 4]]
shallow = copy.copy(original)

shallow[2][0] = 99  
print("Original:", original)
print("Shallow :", shallow)
```

```python
#### deep copy 
##changes in Deep copy  does NOT affect original
original = [1, 2, [3, 4]]
deep = copy.deepcopy(original)
deep[2][0] = 99
print("Original:", original)
print("Deep    :", deep)

###we use when we want to modify new data without affecting original
```
