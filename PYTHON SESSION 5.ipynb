```python
##### map 
### map() applies a function to every element in an iterable and returns a map object.
def square(x):
    return x*x

s = [1, 2, 3, 4]

result = map(square, s)
print(list(result))
```

```python
##### ----reduce()
###-----reduce() is a function in Python that:
##Repeatedly applies a function
##Reduces a list to one single value
###Sum of all numbers
###Product of numbers
##Find max/min
##Combine items one by one
from functools import reduce
def add(a, b):
    return a + b
numbers = [1, 20, 3, 40]
result = reduce(add, numbers)
print(result)
```

```python
numbers = [10, 20, 30, 40]

result = reduce(lambda a, b: a + b, numbers)
print(result)
```

```python
s = [2, 3, 4]

result = reduce(lambda a, b: a * b, s)
print(result)
```
